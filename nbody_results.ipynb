```python
%matplotlib inline

import matplotlib.pyplot as plt 
import numpy as np 
nbodies = 4 
vals = np.loadtxt('results.txt',delimiter=",")

vals = vals.reshape((vals.shape[0]/nbodies,4,nbodies))
# print(vals[0])
pl1x = vals[:,0,0]
pl1y = vals[:,0,2]
pl2x = vals[:,1,0]
pl2y = vals[:,1,2]

fig = plt.figure();
ax = fig.add_subplot(111);
ax.plot(pl1x,pl1y,c='k');
ax.plot(pl2x,pl2y,c='r');
ax.plot(vals[:,2,0],vals[:,2,2],'g')
ax.plot(vals[:,3,0],vals[:,3,2],'b')
```
